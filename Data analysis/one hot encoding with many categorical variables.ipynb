```python
def one_hot_encode(df,variable):
    #first we need to count top 10 variable
    top_10 = [x for x in df[variable].value_counts().sort_values(ascending=False).head(10).index]
    #then we need to add the columns with the dataframe
    for label in top_10:
        df[variable+'_'+label] = np.where(df[variable] == label ,1,0)

data = pd.read_csv('Dataset/train.csv',usecols = ['X1','X2','X3','X4','X5','X6'])

for i in data.columns:
    one_hot_encode(data,i)
data
```
