```
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sns
import pandas as pd
import numpy as
 np


url='https://raw.githubusercontent.com/fazlyrabbi77/Climate-Science/master/Dhaka%20weather%201953-2016.csv'
df = pd.read_csv(url)
df
```

```
#NULL value check
df.isnull().values.any()
```

```
#check all the statistics
df.describe()
```

```
correlation = df.corr()
plt.figure(figsize=(16,8))
sns.heatmap(correlation,annot=True,linewidth=0,vmin =-1,cmap = "RdBu_r")
plt.show
```

```
rain = df.groupby('Month')['Rainfall'].mean()
rain.plot(kind='bar')
```

```
#distribution plot
sns.distplot(rain)
```

```
#find the highest rainfall month
df.groupby('Month')['Rainfall'].mean().sort_values(ascending = False)
```

```
#boxplot of ALL data set
plt.figure(figsize=(16, 6))
ax = sns.boxplot(data=df, orient="h", palette="Set2")
```

```
#we can not predict if all the boxplot are shown all-together
#so we draw each set box plot separately

ax= sns.boxplot(x=df["Rainfall"])
```

```
#if we want to compare two variable relation boxplot can be handful
#here we will plot rel_humidity vs rainfall graph to see the relation between them

plt.figure(figsize=(16, 6)) #this is just to increase the size
ax =sns.boxplot(x="RelativeHumidity",y = "Rainfall", data =df)
```

```
#we can do catagorical boxplot of two variables

plt.figure(figsize=(16,6))
ax=sns.catplot(x="Month",y="Rainfall",kind="box", data =df)
```
