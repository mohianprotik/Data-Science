```
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/fazlyrabbi77/DataProcessing/master/real-estate.csv')
df
```

```
#Renaming columns with inplace
df.rename(columns ={"NUM_BEDROOMS": "BEDROOMS", "NUM_BATH":"BATH"},inplace =True)

```

```
#Checking is there any NAN value
df.isnull().values.any() 
#checking the number of NAN values
df.isnull().values.sum()

```

```
#Showing all the NULL Values as per attribute
null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```
#to identify specific columns NAN value
print(df[df["OWN_OCCUPIED"].isnull()][null_columns])
```

```
#to identify all the columns containing NAN
print(df[df.isnull().any(axis=1)][null_columns])
```

```
#Row wise data filling for ST_NUM column
df.loc[2,'ST_NUM']=197
df.loc[6,'ST_NUM']=208
df
```

```
#now we will remove 12 and replace with NAN for OWN_OCCUPIED columns


cnt=0
for row in df['OWN_OCCUPIED']:
    try:
        int(row)  #only for int value this line will be true and the next line code will execute otherwise for all it will call the exception
        df.loc[cnt, 'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    cnt+=1
df
```

```

#convert all the unwanted things to NAN
df['BEDROOMS'] = pd.to_numeric(df['BEDROOMS'], errors='coerce')
df['BATH'] = pd.to_numeric(df['BATH'], errors='coerce')
df['SQ_FT'] = pd.to_numeric(df['SQ_FT'], errors='coerce')
df
```

```
df['OWN_OCCUPIED'].fillna(df['OWN_OCCUPIED'].mode()[0] ,inplace =True)
df
```

```
#apply mode groupby function
df.groupby('BEDROOMS')['SQ_FT'].median()
```

```
df['SQ_FT'] = df['SQ_FT'].fillna(df.groupby('BEDROOMS')['SQ_FT'].transform('median'))
df
```

```
df.loc[2,'BEDROOMS']=1
df.loc[5,'BEDROOMS']=1
df.loc[8,'BEDROOMS']=3
```

```
#bfill will fill the value of NAN with next colums value
#ffill will fill the value of NAN with previous colums value
df['BATH']= df['BATH'].fillna(method='ffill')
df
```

```
df.loc[4,'PID']=100005000.0
```

```
#Converting street number to int
df.PID = df.PID.astype('int64') 
df.ST_NUM = df.ST_NUM.astype('int64') 
df.BEDROOMS = df.BEDROOMS.astype('int64') 
df.BATH = df.BATH.astype('int64') 
df.SQ_FT = df.SQ_FT.astype('int64') 
df
```

DATA TRANSFORMATION

```
#for converting the ST_NAME to number 
#Label Encoding for catagorical data process

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['ST_NAME'] = le.fit_transform(df['ST_NAME'])
df
```

```
#we can do the same thing with mapping but here we need to mention every mapped value
#use of mapping function
mapping = {'Y':1,'N':2}
df['OWN_OCCUPIED'] = df['OWN_OCCUPIED'].map(mapping)
df
```

```
#one hot encoder for nominal data
#it will divide a columns into several columns depends on the how many catagorical data present in the col
#like OWN_OCCUPIED has 2 types of value 1 and 2 so it will divide OWN_OCCUPIED in two different columns and fill the value with bollean 0/1

df1= pd.get_dummies(df,columns=['OWN_OCCUPIED'])
df1
```

```
#now we want to keep all data in a certain range of value
#for this we will use MinMaxScaler function it uses the eqaution of
#xnew=x-min(x)/max(x)-min(x) 

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df)
scaled = pd.DataFrame(scaler.transform(df),columns=df.columns)
scaled
```
